```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importing dataset 
df=pd.read_csv('BigMart_data.csv')
# droping the null values
df=df.dropna(how='any')
# view the top 5 data rows as
df.head()
```

```python
#3 Line Chart
#  what is the mean price per item ??
price_by_item = df.groupby('Item_Type').Item_MRP.mean()[:10]
price_by_item
```

```python
# mean price based on each item type
price_by_item = df.groupby('Item_Type').Item_MRP.mean()[:10]
# basically storing the name/item_types and values of the items  in x and y by forming list
x= price_by_item.index.tolist()
y =price_by_item.values.tolist()

# set the figure size
plt.figure(figsize=(14,8))
# title of the chart
plt.title('Mean price for each item type')
# setting the labels
plt.xlabel('Item Type')
plt.ylabel('MRP Price')
# adding ticks 
plt.xticks(labels=x,ticks=np.arange(len(x)))
# plot the line chart
plt.plot(x,y)
```

```python
# 4 bar charts
# what is the mean sales for each outlet size??
```

```python
# sales by outlet size
sales_by_outlet_size = df.groupby('Outlet_Size').Item_Outlet_Sales.mean()
# sorting the numbers by sales
sales_by_outlet_size.sort_values(inplace=True)
#  creating x and y variables to store name and values 
x= sales_by_outlet_size.index.tolist()
y= sales_by_outlet_size.values.tolist()

# set the axis labels
plt.xlabel('Outlet Size')
plt.ylabel('Sales')
# set title
plt.title('Mean sales for each outlet size')

# set xticks
plt.xticks(labels=x,ticks=np.arange(len(x)))
# plot the bar chart
plt.bar(x,y,color=('red','blue','magenta'))
```

```python
# 5 HISTOGRAM
# histogram- it used to undersant the DISTRIBUTION of Item price
# Histograms are very common type of plots when we are looking at data like height & weight and stock prices,etc which has countinous time in nature
#  Histrogram data is plotted within the range against its frequency
# hitogram are very commonposstion graphs in probability and statictics and form the basis of various distributions like normal distribution , tidistributions,etc
# ****************\what is the distribution of the item price?? /**************
```

```python
# title
plt.title('Item MRP (price) distribution ')
#  xlabel
plt.xlabel('Item_MRP')
# ylabel
plt.ylabel('Frequency')
# plot histogram bins 20 is basically the range that we want here for each block
plt.hist(df['Item_MRP'],bins=20,color='lightblue');
```

```python
#6 box plot - HELPS IN UNDERSTANDING THE OUTLIERS IN DATA AND DISTRIBUTION OF DATA
#  what is the distribution of slaes usinng box plots
# box plot shows the three quartile values of the distribution along with extreme values.
#  The" whiskers" extend to the points that lie within 1.5 IQRs of the lower and upper quartile and then observations outside this range are displayed independently
#  This means that each value in boxplot corresponds to actual observations in thedata.
# try to visualize thr distribution of item_Outlet_Sales of items.
```

```python
data=df[['Item_Outlet_Sales']]
# create outlier point shape
red_diamond=dict(markerfacecolor='r',marker='D')
# set title
plt.title('Item Sales distribution')
# make the boxplot
plt.boxplot(data.values, labels=['Item Sales'], flierprops=red_diamond);
```

```python
#  MULTIPLE BOX PLOTS IN SINGLE FRAME
data=df[['Item_Weight','Item_MRP']]

# create outlier point shape
red_diamond = dict(markerfacecolor='r',marker='D')

# genrate suplots
fig, ax = plt.subplots()

# make the boxplot
plt.boxplot(data.values,labels=['Item_Weight','Item_MRP']);
```

```python
# 7 Violin Plots - It tell us Density Distribution of Item Weights And Item Price
```

```python
data=df[['Item_Weight','Item_MRP']]

# genrate subplots
fig,ax=plt.subplots()

# add label to x axis
plt.xticks(ticks=[1,2],labels=['Item Weight','Item MRP'])

# make the violetplot
plt.violinplot(data.values);
```
