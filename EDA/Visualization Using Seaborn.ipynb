```python
# VISUALIZATION WITH SEABORN 
# SEABORN IS A PYTHON DATA VISUALIZATION Library based on Matplotlib
# It provied a high level interface for drawing atttractive and informative statistical graphics
# It provide choices for plot style and color defaults,defines samle high level functions statistical plot types and integrates with the functionality provided by Pandas DataFrames
# The main idea of seaborn is that it provides high level commands to create a variety of plot types usefuul for statistical data exploration ,and even more statistical model training
```

```python
# importing required libraries
import seaborn as sns
sns.set()
sns.set(style="darkgrid")

import pandas as pd
import numpy as np

# importing matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
# read the dataset
df=pd.read_csv('BigMart_data.csv')
# drop the null values
df=df.dropna(how='any')
# multiply the Item Visibility by 100 to increase size
df["Visibility_Scaled"]= df["Item_Visibility"]*100
# view the data top 5 heads
df.head()
```

```python
# no1. -LINE CHART
sns.lineplot(x="Item_Weight",y="Item_MRP", data=df[:50])
```

```python
# no.2 - BAR Chart
sns.barplot(x="Item_Type",y="Item_MRP",data=df[:5])
```

```python
# no.3 -HISTOGRAM - in this all we need is to pass plot name in distplot function
sns.distplot(df['Item_MRP'])
```

```python
# 4- Boxplots
sns.boxplot(df['Item_Outlet_Sales'],orient='horizontal')
```

```python
# no.5 -VIOLIN PLOT
sns.violinplot(df['Item_Outlet_Sales'], orient='hortizental', color='magenta')
```

```python
# no -Scatterplot by using relplot func 
sns.relplot(x='Item_MRP',y="Item_Outlet_Sales", data=df[:200], kind="scatter")
```

```python
#  no - Hue Semantic it allows us to add another dimensions to the plot by coloring the points accoding to a third variable
# therefor in seaborn it is referd as hue semantic
sns.relplot(x="Item_MRP",y="Item_Outlet_Sales",hue="Item_Type",data=df[:150])
```

```python
#  Applying hue semantic in Lineplots for different categories of the outlet size
sns.lineplot(x="Item_Weight",y="Item_MRP",hue="Outlet_Size",data=df[:150])
```

```python
# no - bubble plot
sns.relplot(x="Item_MRP",y="Item_Outlet_Sales", data=df[:200] , kind="scatter",size="Visibility_Scaled",hue="Visibility_Scaled");
```

```python
# CATEGORY WISE SUB PLOT
# we can actually create plot using Categories in seaborn
#  we have created scatter plots dor each Outlet_SIze'
sns.relplot(x="Item_Weight",y="Item_Visibility",hue='Outlet_Size',style='Outlet_Size',col='Outlet_Size',data=df[:150],kind="scatter")
```

```python
# ADVANCED CATEGORICAL PLOTS IN SEABORN
# 3 DIFFERENT TYPES OF FAMILIES IN SEABORN'
# 1) Categorical Scatterplots:
# stripplot()(with kind='strip',the default)
# swarmplot()(with kind='swarm')
# 2). Categorical distribution plots
# boxplot()(with kind='box')
# violinplot()(with kind='violin')
# 3) Categorical estimate plots
# barplot(with kind='bar')
```

```python
sns.catplot(x="Outlet_Size",y="Item_Outlet_Sales",kind='strip',data=df[:250])
```

```python
sns.catplot(x="Outlet_Size",y="Item_Outlet_Sales",kind="swarm",data=df[:250])
```

```python
# difference between the stripand swam os that in swarm plot the data points are not overlapping whereas in strip plot the data variables are overlapping each other.
# but the swarm plot does always scare welll for large no. of data stes
# 
```

```python
# 2).Categorial Distribution Plots using box plot
sns.catplot(x="Outlet_Size",y="Item_Outlet_Sales",kind="box",data=df[:250])
```

```python
sns.catplot(x="Outlet_Size",y="Item_Outlet_Sales",kind="violin",data=df[:250])
```

```python
sns.catplot(x="Outlet_Size",y="Item_Outlet_Sales",kind="bar",data=df[:250])
```

```python
# DESNITY PLOTS
# rATHER THAN A HISTOGRAM , WE CAN GET A SMOOTH ESTIMATE OF THE DISTRIBUTION USING DENSITY PLOT,WHICH SEABORN DOES WITH sns.kdeplot
# thess help us get a smooth estimate of distribution using a kernel density estimation which
```

```python
# distribution of the item visibility variable
plt.figure(figsize=(7,7))
sns.kdeplot(df['Item_Visibility'],shade=True);
```

```python
# distribution of Item MRP
plt.figure(figsize=(7,7))
sns.kdeplot(df['Item_MRP'],shade=True);
```

```python
# HISTOGRAM AND DENSITY PLOT using SEABORN
# Histograms and KDE can be Combineed through distplot
```

```python
plt.figure(figsize=(7,7))
sns.distplot(df['Item_Outlet_Sales']);
```

```python
plt.figure(figsize=(7,7))
sns.displot(df['Item_Outlet_Sales']);
```
